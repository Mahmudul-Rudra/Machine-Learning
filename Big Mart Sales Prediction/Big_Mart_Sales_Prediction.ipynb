Data Pre-Processing

```python
big_mart_data.head()
```

```python
big_mart_data['Item_Fat_Content'].value_counts()
```

```python
big_mart_data.replace({'Item_Fat_Content': {'low fat':'Low Fat','LF':'Low Fat', 'reg':'Regular'}}, inplace=True)
```

```python
big_mart_data['Item_Fat_Content'].value_counts()
```

Label Encoding

```python
encoder = LabelEncoder()
```

```python
big_mart_data['Item_Identifier'] = encoder.fit_transform(big_mart_data['Item_Identifier'])

big_mart_data['Item_Fat_Content'] = encoder.fit_transform(big_mart_data['Item_Fat_Content'])

big_mart_data['Item_Type'] = encoder.fit_transform(big_mart_data['Item_Type'])

big_mart_data['Outlet_Identifier'] = encoder.fit_transform(big_mart_data['Outlet_Identifier'])

big_mart_data['Outlet_Size'] = encoder.fit_transform(big_mart_data['Outlet_Size'])

big_mart_data['Outlet_Location_Type'] = encoder.fit_transform(big_mart_data['Outlet_Location_Type'])

big_mart_data['Outlet_Type'] = encoder.fit_transform(big_mart_data['Outlet_Type'])
```

```python
big_mart_data.head()
```

Splitting features and Target

```python
X = big_mart_data.drop(columns='Item_Outlet_Sales', axis=1)
Y = big_mart_data['Item_Outlet_Sales']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into Training data & Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Machine Learning Model Training

XGBoost Regressor

```python
regressor = XGBRegressor()
```

```python
regressor.fit(X_train, Y_train)
```

Evaluation

```python
training_data_prediction = regressor.predict(X_train)
test_data_prediction = regressor.predict(X_test)
```

```python
#Training Data visualization
plt.figure(figsize=(50, 50))
sns.regplot(x=Y_train, y=training_data_prediction, scatter_kws={'color': 'red'})
```

```python
#Testing Data visualization
plt.figure(figsize=(50, 50))
sns.regplot(x=Y_test, y=test_data_prediction, scatter_kws={'color': 'red'})
```

```python
r2_train = metrics.r2_score(Y_train, training_data_prediction)*100
r2_test = metrics.r2_score(Y_test, test_data_prediction)*100
cross_validation_score = (cross_val_score(regressor, X, Y, cv=5, scoring ='neg_mean_squared_error').mean())
```

```python
print("R2_score of XGBRegression Model on Training Data: {:.2f}".format(r2_train))
print("R2_score of XGBRegression Model on Testing Data: {:.2f}".format(r2_test))
print("Cross validation Score: {:.2f}".format(cross_validation_score))
```

#Linear Regression Model

```python
linear_regressor = LinearRegression()
linear_regressor.fit(X_train, Y_train)
```

```python
training_data_prediction_linear = linear_regressor.predict(X_train)
testing_data_prediction_linear = linear_regressor.predict(X_test)
```

```python
#Training Data visualization
plt.figure(figsize=(50, 50))
sns.regplot(x=Y_train, y=training_data_prediction_linear, scatter_kws={'color': 'red'})
```

```python
#Testing Data visualization
plt.figure(figsize=(50, 50))
sns.regplot(x=Y_test, y=testing_data_prediction_linear, scatter_kws={'color': 'red'})
```

```python
r2_train_linear = metrics.r2_score(Y_train, training_data_prediction_linear)*100
r2_test_linear = metrics.r2_score(Y_test, testing_data_prediction_linear)*100
cross_validation_score_linear = (cross_val_score(linear_regressor, X, Y, cv=5, scoring ='neg_mean_squared_error').mean())
```

```python
print("R2_score of Linear Regression Model on Training Data: {:.2f}".format(r2_train_linear))
print("R2_score of Linear Regression Model on Testing Data: {:.2f}".format(r2_test_linear))
print("Cross validation Score: {:.2f}".format(cross_validation_score_linear))
```
